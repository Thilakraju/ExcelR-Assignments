```python
user_item_matrix = df.pivot_table(values='Book_Rating', index='User_id', columns='Book_Title', aggfunc='mean', fill_value=None, margins=False, dropna=True, margins_name='All', observed=False, sort=True)
```

```python
user_item_matrix
```

```python
user_item_matrix.index = df.User_id.unique()
```

```python
user_item_matrix
```

```python
user_item_matrix.fillna(0, inplace=True)
```

```python
user_item_matrix
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_item_matrix.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df.index = df.User_id.unique()
user_sim_df.columns =df.User_id.unique()
```
