```python
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import re
import nltk
```

```python
with open("Elon_musk.csv","r",encoding="ISO-8859-1")as input:
    Elon_reviews=input.read()
type(Elon_reviews)
Elon_reviews
```

```python
el_rev_string=Elon_reviews
el_rev_string=re.sub("^[A-Za-z]+"," ",el_rev_string).lower()
el_rev_words=el_rev_string.split(" ")
```

```python
ret=re.sub("[^A-Za-z]+"," ",'abc@Xy8z')
```

```python
ret
```

```python
#nltk.download('stopwords')
```

```python
stop_words=stopwords.words('english')
```

```python
el_rev_words=[w for w in el_rev_words if not w in stop_words]
```

```python
el_rev_string=" ".join(el_rev_words)
```

```python
el_rev_string
```

```python
wordcloud_el=WordCloud(background_color="black",max_words=100).generate(el_rev_string)
plt.imshow(wordcloud_el)
plt.axis("off")
plt.show()
```

```python
with open("positive-words.txt","r")as pos:
    poswords=pos.read().split("\n")
```

```python
print(poswords)
```

```python
with open("negative-words.txt","r")as neg:
    negwords=neg.read().split("\n")
```

```python
print(negwords)
```

```python
el_neg_in_neg=" ".join([w for w in el_rev_words if not w in negwords])
```

```python
wordcloud_neg_in_neg=WordCloud(background_color="black",width=1800,height=1400).generate(el_neg_in_neg)
plt.imshow(wordcloud_neg_in_neg)
plt.axis("off")
plt.show()
```

```python
el_pos_in_pos=" ".join([w for w in el_rev_words if not w in poswords])
```

```python
wordcloud_pos_in_pos=WordCloud(background_color="black",width=1800,height=1400).generate(el_pos_in_pos)
plt.imshow(wordcloud_pos_in_pos)
plt.axis("off")
plt.show()
```
